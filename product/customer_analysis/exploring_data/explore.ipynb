```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('../../../../dataset.csv')
df = df.dropna()

df = df.drop(columns=['Row ID', 'Order ID'])

df
```

# Which categories perform the best?

```python

df['Profit Markup'] = df['Profit'] / df['Sales']


sum_of_categories = df.groupby('Category')['Profit'].sum()
mean_of_categories = df.groupby('Category')['Profit'].mean()
markup_of_categories = df.groupby('Category')['Profit Markup'].mean()



print("Top 5 Categories with highest sum of profit:")
print(sum_of_categories)

print("\nTop 5 Categories with highest mean profit:")
print(mean_of_categories)

print("\nTop 5 Categories with highest profit markup:")
print(markup_of_categories)

```

# Sub categories?

```python

df['Profit Markup'] = df['Profit'] / df['Sales']


sum_of_subcategories = df.groupby('Sub-Category')['Profit'].sum().reset_index()
mean_of_subcategories = df.groupby('Sub-Category')['Profit'].mean().reset_index()
markup_of_subcategories = df.groupby('Sub-Category')['Profit Markup'].mean().reset_index()


top5_sum = sum_of_subcategories.nlargest(5, 'Profit')
top5_mean = mean_of_subcategories.nlargest(5, 'Profit')
top5_markup = markup_of_subcategories.nlargest(5, 'Profit Markup')

print("Top 5 Sub-Categories with highest sum of profit:")
print(top5_sum)

print("\nTop 5 Sub-Categories with highest mean profit:")
print(top5_mean)

print("\nTop 5 Sub-Categories with highest profit markup:")
print(top5_markup)
```

# Which prodicts perform the best?

```python
df_names = df[['Product ID', 'Product Name']].drop_duplicates()

df['Profit Markup'] = df['Profit'] / df['Sales']


sum_of_categories = df.groupby('Product ID')['Profit'].sum().reset_index()
mean_of_categories = df.groupby('Product ID')['Profit'].mean().reset_index()
markup_of_categories = df.groupby('Product ID')['Profit Markup'].mean().reset_index()


sum_with_names = sum_of_categories.merge(df_names, on='Product ID')
mean_with_names = mean_of_categories.merge(df_names, on='Product ID')
markup_with_names = markup_of_categories.merge(df_names, on='Product ID')

# Get top 5
top5_sum = sum_with_names.nlargest(5, 'Profit')
top5_mean = mean_with_names.nlargest(5, 'Profit')
top5_markup = markup_with_names.nlargest(5, 'Profit Markup')

print("Top 5 products with highest sum of profit:")
print(top5_sum)

print("\nTop 5 products with highest mean profit:")
print(top5_mean)

print("\nTop 5 products with highest profit markup:")
print(top5_markup)


```
